```python
from sklearn.model_selection import train_test_split , GridSearchCV
from sklearn.preprocessing import LabelEncoder

from sklearn.linear_model import LinearRegression,Ridge
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
df.describe(include="object").columns
```

```python
features=['Hours_Studied', 'Attendance', 'Sleep_Hours', 'Previous_Scores',
       'Tutoring_Sessions', 'Physical_Activity','Gender',
       'Extracurricular_Activities']
target="Exam_Score"
```

```python
df_model=df[features+[target]].copy()
df_model
```

```python


le = LabelEncoder()
df_model["Extracurricular_Activities"] = le.fit_transform(df_model["Extracurricular_Activities"])
df_model
```

```python
le = LabelEncoder()
df_model["Gender"] = le.fit_transform(df_model["Gender"])
df_model
```

```python
X=df_model[features]
Y=df_model[target]
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
len(Y_test)
len(Y_train)
```

```python
models = {
    "LinearRegression": {
        "model": LinearRegression(),
        "params": {}
    },
    "DecisionTree": {
        "model": DecisionTreeRegressor(),
        "params": {"max_depth": [3, 5, 10], "min_samples_split": [2, 5]}
    },
    "RandomForest": {
        "model": RandomForestRegressor(),
        "params": {"n_estimators": [50, 100], "max_depth": [5, 10]}
    }
}

```

```python
best_models = []
```

```python

for name, config in models.items():
    print(f"Training {name}")

    grid=GridSearchCV(config["model"],config["params"],cv=5,scoring="neg_mean_squared_error")
    grid.fit(X_train,Y_train)

    Y_pred=grid.predict(X_test)
    rmse=np.sqrt(mean_squared_error(Y_test,Y_pred))
    r2=r2_score(Y_test,Y_pred)

    best_models.append({
        "model":name,
        "best_params" : grid.best_params_,
        "rmse" :rmse,
        "r2" :r2

     })
```

```python
best_models
```

```python
results_df=pd.DataFrame(best_models)
```

```python
results_df.sort_values(by="rmse")
```

```python
import joblib
best_row=results_df.sort_values(by="rmse").iloc[0]
best_row
```

```python
best_model_name=best_row["model"]
best_model_name
```

```python
best_model_config=models[best_model_name]
best_model_config
```

```python
final_model=best_model_config["model"]
```

```python
final_model.fit(X,Y)
```

```python
joblib.dump(final_model,"best_model.pkl")
```
